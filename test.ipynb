```python
import pandas as pd

# Carrega o dataset
df=pd.read_csv('C:/trabalhodados/uber.csv')
#df = pd.read_csv('C:/Banco de dados/uber.csv', encoding='cp1252')

# Exibe as 5 primeiras linhas do DataFrame
print("Primeiras 5 linhas do DataFrame:")
display(df.head())

# Exibe informações gerais sobre o DataFrame, como tipos de dados e valores nulos
print("\nInformações sobre o DataFrame:")
df.info()
df['Ride Distance'] = df['Ride Distance'].fillna(0)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Replace missing values with 0 in columns: 'Avg VTAT', 'Avg CTAT'
    df = df.fillna({'Avg VTAT': 0, 'Avg CTAT': 0})
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
# Converte a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

print("Tipo da coluna 'Date' após a conversão:", df['Date'].dtype)
```

```python
# Calcula a média da coluna 'Driver Ratings', ignorando valores nulos
media_avaliacoes_motoristas = df['Driver Ratings'].mean()

# Preenche os valores nulos com a média calculada
df['Driver Ratings'].fillna(media_avaliacoes_motoristas, inplace=True)

print("Número de valores nulos em 'Driver Ratings' após o preenchimento:", df['Driver Ratings'].isnull().sum())
```

```python
# Ajusta os valores nulos na coluna 'Ride Distance' para 0
df['Ride Distance'].fillna(9999, inplace=True)

# Depois do ajuste, vamos verificar novamente o número de valores nulos
print("Valores nulos em 'Ride Distance' após o ajuste:", df['Ride Distance'].isnull().sum())

# Exibe as primeiras 10 linhas para visualizar as mudanças (se houver valores nulos nas primeiras linhas)
print("\nPrimeiras linhas do DataFrame após o ajuste:")
df.head(10)
```

```python
# Filtra o DataFrame para incluir apenas viagens com status 'Completed'
completed_rides = df[df['Booking Status'] == 'Completed']

print(f"Total de viagens concluídas: {len(completed_rides)} de {len(df)}")
display(completed_rides.head())
```

```python
# Filtra o DataFrame usando múltiplas condições combinadas com o operador '&'
long_upi_rides = df[(df['Payment Method'] == 'UPI') & (df['Ride Distance'] > 20)]

print(f"Total de viagens longas (> 20 km) pagas com UPI: {len(long_upi_rides)}")
display(long_upi_rides.head())
```

```python
# Agrupa por 'Vehicle Type' e conta o número de ocorrências em cada grupo
rides_by_vehicle_type = df.groupby('Vehicle Type').size().reset_index(name='Total Rides')

print("Contagem de viagens por tipo de veículo:")
display(rides_by_vehicle_type)
```

```python
# Agrupa por 'Payment Method' e calcula a média do 'Booking Value'
avg_booking_value_by_payment = df.groupby('Payment Method')['Booking Value'].mean().reset_index()

print("Valor médio da viagem por método de pagamento:")
display(avg_booking_value_by_payment)
```

```python
# Agrupa por 'Booking Status' e calcula a soma da 'Ride Distance'
total_distance_by_status = df.groupby('Booking Status')['Ride Distance'].sum().reset_index()

print("Distância total percorrida por status da viagem:")
display(total_distance_by_status)
```

```python
# Usa a função `agg` para aplicar múltiplas funções de agregação
summary_by_vehicle = df.groupby('Vehicle Type')['Booking Value'].agg(['mean', 'median', 'sum', 'count'])

print("Resumo estatístico do valor da viagem por tipo de veículo:")
display(summary_by_vehicle)
```

```python
# Usa `value_counts()` para contar a frequência de cada valor único
booking_status_counts = df['Booking Status'].value_counts().reset_index()
booking_status_counts.columns = ['Booking Status', 'Count']

print("Frequência de cada status de reserva:")
display(booking_status_counts)
```

```python
# Filtra as viagens onde o cancelamento foi feito pelo cliente (valor 1)
cancelled_by_customer_count = df['Cancelled Rides by Customer'].sum()

print(f"Total de viagens canceladas por clientes: {int(cancelled_by_customer_count)}")
```

```python
# Conta a frequência das razões de cancelamento
cancellation_reasons = df['Reason for cancelling by Customer'].value_counts().reset_index()
cancellation_reasons.columns = ['Reason', 'Count']

print("Razões mais comuns para cancelamento por clientes:")
display(cancellation_reasons)
```

```python
from sqlalchemy import create_engine
import psycopg2 # Necessário para a conexão, mesmo que usado pela SQLAlchemy

# --- Passo 1: Configurar a conexão com o banco de dados PostgreSQL ---
# ATENÇÃO: Substitua 'seu_usuario', 'sua_senha', 'seu_host', 'seu_banco' pelos seus dados reais
DB_USER = 'postgres'
DB_PASSWORD = '123'
DB_HOST = 'localhost' # Ou o endereço IP do seu servidor PostgreSQL
DB_PORT = '5432' # A porta padrão do PostgreSQL
DB_NAME = 'cesul-aula'

# A string de conexão (formatada para o PostgreSQL com psycopg2)
#conn_string = f'postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'

engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}",
    connect_args={'options': '-c client_encoding=UTF8'}
)

#engine = create_engine(conn_string)

print("Conexão com o banco de dados PostgreSQL estabelecida com sucesso.")
```

```python
# --- Passo 2: Criar a tabela e inserir os dados do DataFrame ---
# Nome da tabela no banco de dados
table_name = 'viagens_uber'

df.columns = (
    df.columns
        .str.strip()                         # 1. Remove espaços extras no início e no fim dos nomes
        .str.lower()                         # 2. Converte todos os caracteres para minúsculo
        .str.replace(" ", "_")               # 3. Troca espaços por underline (_)
        .str.normalize("NFKD")               # 4. Normaliza acentuação (ex: "ç" → "c", "á" → "a") #NFKD = Normalization Form Compatibility Decomposition
        .str.encode("ascii", errors="ignore")# 5. Converte para ASCII e remove caracteres não suportados
        .str.decode("utf-8")                 # 6. Decodifica de volta para string (utf-8)
)

# O método to_sql() cria a tabela e insere os dados
# - name: nome da tabela
# - con: objeto de conexão
# - if_exists: 'replace' apaga a tabela se ela existir e a recria. Outras opções: 'append' (adiciona linhas) ou 'fail' (gera erro)
# - index=False: para não incluir o índice do DataFrame como uma coluna na tabela

try:
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
    print(f"\nDataFrame inserido com sucesso na tabela '{table_name}'.")

except Exception as e:
    print(f"\nErro ao inserir o DataFrame no banco de dados: {e}")
    # Você pode querer adicionar um `rollback()` em um cenário de transação
```

```python
try:
    cancellation_reasons.to_sql('cancelamentos', con=engine, if_exists='replace', index=False)
    print(f"\nDataFrame inserido com sucesso na tabela '{table_name}'.")

except Exception as e:
    print(f"\nErro ao inserir o DataFrame no banco de dados: {e}")
    # Você pode querer adicionar um `rollback()` em um cenário de transação
```

```python
# --- Passo 3 Fazer uma consulta e ler os dados de volta para um novo DataFrame ---
# Usando pd.read_sql_table() para ler a tabela inteira
try:
    df_from_db = pd.read_sql_table(table_name, con=engine)
    print(f"\nDados lidos com sucesso da tabela '{table_name}':")
    print(df_from_db.head())
    print(f"\nTotal de linhas lidas: {len(df_from_db)}")

except Exception as e:
    print(f"\nErro ao ler a tabela do banco de dados: {e}")


# Fechar a conexão
engine.dispose()
print("\nConexão com o banco de dados fechada.")
```

```python
# Exemplo de consulta usando uma query SQL personalizada com pd.read_sql_query()
query = f"SELECT \"booking_status\", COUNT(*) AS total_viagens FROM {table_name} GROUP BY \"booking_status\""
try:
    df_summary = pd.read_sql_query(query, con=engine)
    print("\nResultado da consulta agregada (Contagem por Booking Status):")
    print(df_summary)

except Exception as e:
    print(f"\nErro ao executar a consulta SQL: {e}")
```
